```python
import pandas as pd
import numpy as np
df = pd.read_csv ("World Population Growth.csv")
df.head()
```

```python
#statistik dasar
df['Populasi'] = df['Population'].str.replace(',','').astype(float)
df['Pertumbuhan'] = df['Yearly Growth %'].str.replace('%','').astype(float)
df['Tahun'] = df['Year'].astype(int)
print("================================================")
mean_pop = df['Populasi'].mean()
median_pop = df['Populasi'].median()
std_pop = df['Populasi'].std()
print(mean_pop)
print(median_pop)
print(std_pop)
print("================================================")
mean_tumbuh = df['Pertumbuhan'].mean()
median_tumbuh = df['Pertumbuhan'].median()
std_tumbuh = df['Pertumbuhan'].std()
print(f"{mean_tumbuh:.2f}%")
print(f"{median_tumbuh:.2f}%")
print(f"{std_tumbuh:.2f}%")
```

```python
#regresi Linear
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
df_clean = df.dropna(subset=['Tahun', 'Populasi'])
x = df_clean[['Tahun']]
y = df_clean['Populasi']
model = LinearRegression()
model.fit(x,y)
intercept = model.intercept_
slop = model.coef_[0]
r2 = r2_score(y, model.predict(x))

print(f"{intercept:.2f} + {slop:.2f} * Year")
print(f"{r2:.4f}")
```

```python
#Visualisasi
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.bar(df_clean['Tahun'], df_clean['Populasi'], color='blue')
plt.title('Populasi per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Populasi')
plt.grid(axis='y', linestyle='--')
plt.show()
```
