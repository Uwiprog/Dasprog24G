```python
#Special Exception -> Own Exception

class Umurexception(Exception) :
    pass

def cekUmur(umur):
    if(umur<=17):
        raise Umurexception("Film dewasa harus di atas 17 tahun")

try:
    cekUmur(15)
    print("Selamat datang")
except Umurexception as ue:
    print(ue)

else:
    print("Pilih film sesuai umur anda")

finally:
    print("program selesai!..")
```

```python
import pandas as pd
import numpy as np
#Library Visual -> Matplotib & SEaborn
import matplotlib.pyplot as plt

np.random.seed(42)

bulan = ['Jan','Feb','Mar','Apr','Mei','Jun','Jul','Aug','Sep','Oct','Nov','Des']
Minggu1 = np.random.randint(200000,500001,12)
Minggu2 = np.random.randint(200000,500001,12)
Minggu3 = np.random.randint(200000,500001,12)
Minggu4 = np.random.randint(200000,500001,12)

data = {
    'Bulan' : bulan,
    'Minggu_1': Minggu1,
    'Minggu_2': Minggu2,
    'Minggu_3': Minggu3,
    'Minggu_4': Minggu4,
}

df = pd.DataFrame(data)

df['Total_Pengeluaran'] = df[['Minggu_1','Minggu_2','Minggu_3','Minggu_4']].sum(axis=1)

df.head()
```

```python
#Diagram Bar -> proses perbandingan
#Buat canvas
plt.figure(figsize=(12,6)) #width,height -> inch
plt.bar(df['Bulan'],df['Total_Pengeluaran'], color='skyblue')
plt.title("Pengeluaran Mahasiswa per tahun")
plt.xticks(rotation=-45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
plt. figure(figsize=(12,8))

for minggu in ['Minggu_1','Minggu_2','Minggu_3','Minggu_4']:
    plt.plot(df['Bulan'],df[minggu], marker='o', label=minggu)

plt.title("Trend pengeluaran mahasiswa perminggu")
plt.xlabel("Bulan")
plt.ylabel("Trend pengeluaran mahasiswa")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#pie diagram
plt.figure(figsize=(8,8))
plt.pie(df['Total_Pengeluaran'],labels=df['Bulan'],autopct='%1.1f%%')
plt.show()
```

```python
#regresi Linear
#Open source -> scikit -> sklearn
from sklearn.linear_model import LinearRegression

df['Bulan_Angka'] = np.arange(1,13)

x = df[['Bulan_Angka']]
y = df['Total_Pengeluaran']

model = LinearRegression()
model.fit(x,y)

bulan_13 = model.predict([[13]])[0]
bulan_14 = model.predict([[14]])[0]

print(f"Prediksi Pengeluaran Mahasiswa Januari 2026:{bulan_13:,.0f}")
print(f"Prediksi Pengeluaran Mahasiswa Januari 2026:{bulan_14:,.0f}")
```
