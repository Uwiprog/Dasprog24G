```python
angka1 = int(input("Angka 1:"))
angka2 = int(input("Angka 2:"))
try :
    hasil = angka1 / angka2
except ZeroDivisionError:
    print("Tidak boleh pembagian dengan angka 0")
else:
    print(f"hasil pembagian {angka1} dan {angka2} = {hasil}")
finally:
    print("Pemrograman selesai!")
```

```python
try :
    angka1 = int(input("Angka 1:"))
    angka2 = int(input("Angka 2:"))
    hasil = angka1 / angka2
except ValueError:
    print("input harus angka")
except ZeroDivisionError:
    print("Tidak boleh pembagian dengan angka 0")
else:
    print(f"hasil pembagian {angka1} dan {angka2} = {hasil}")
finally:
    print("Pemrograman selesai!")
```

```python
try :
    angka1 = int(input("Angka 1:"))
    angka2 = int(input("Angka 2:"))
    hasil = angka1 / angka2
except (ValueError, ZeroDivisionError) as err:
    print(err)
else:
    print(f"hasil pembagian {angka1} dan {angka2} = {hasil}")
finally:
    print("Pemrograman selesai!")
```

```python
#indexError

listData = ['1','2','3','4']

try:
    print(listData[6])
except IndexError as error :
    print(error)
finally:
    print("Selesai!..")
```

```python
try:
    TotalData = int(input('Jumlah data yang akan dimasukkan: '))
    listNama = []
    for i in range(0,TotalData):
        nama = input("Isikan nama siswa:")
        listNama.append(nama)
    indeks = int(input("No Urut berapa yang akan ditampilkan?"))
    print(f"Nama siswa ke {indeks} : {listNama[indeks-1]}")
except (ValueError,IndexError) as error:
    print(error)
else:
    print(listNama)
finally:
    print("Program Selesai!..")
```

```python
#key Error
data = {
   'day' : ['mon','tue','wed','thu','fri','sat','sun']   
}
try:
    data['month']
except KeyError :
    print("Tidak ada atribut yang sama")
else:
    print(data)
finally:
    print("Selesai")
```

```python
#Special Exception -> Own Exception

class Umurexception(Exception) :
    pass

def cekUmur(umur):
    if(umur<=17):
        raise Umurexception("Film dewasa harus di atas 17 tahun")

try:
    cekUmur(15)
    print("Selamat datang")
except Umurexception as ue:
    print(ue)

else:
    print("Pilih film sesuai umur anda")

finally:
    print("program selesai!..")
```
